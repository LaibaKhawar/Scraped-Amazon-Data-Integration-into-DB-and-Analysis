```python
def clean_average_review(average_review):
    if isinstance(average_review, str):
        # Remove ' ratings' from the string
        cleaned_review = average_review.replace(' ratings', '')
        # Split the string by comma and take the first part
        cleaned_review = cleaned_review.split(',')[0]
        return cleaned_review
    else:
        return average_review


# Apply the cleaning function to the 'Average_Review' column
df['Average_Review'] = df['Average_Review'].apply(clean_average_review)

# Display the DataFrame
print(df['Average_Review'])
```

```python
def clean_brand_name(brand_name):
    if isinstance(brand_name, str):
        # Remove ' store' from the string
        cleaned_name = brand_name.replace(' store', '')
        # Remove 'brand: ' if it exists
        cleaned_name = cleaned_name.replace('brand: ', '')
        return cleaned_name
    else:
        return brand_name

# Apply the cleaning function to the 'Brand_Name' column
df['Brand_Name'] = df['Brand_Name'].apply(clean_brand_name)

# Display the DataFrame
print(df['Brand_Name'])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def clean_product_detail(text):
    if isinstance(text, str):
        # Tokenize the text
        tokens = word_tokenize(text)
        # Remove punctuation and lowercase the tokens
        tokens = [word.lower() for word in tokens if word.isalnum()]
        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if not word in stop_words]
        # Lemmatization
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
        return ' '.join(lemmatized_tokens)
    else:
        return ''
df['Cleaned_Product_Detail'] = df['Product_Detail'].apply(clean_product_detail)


# Check if there are any NaN values in the DataFrame
nan_values_exist = df['Cleaned_Product_Detail'].isna().any()

if nan_values_exist:
    print("There are NaN values in the 'Cleaned_Product_Detail' column.")
else:
    print("There are no NaN values in the 'Cleaned_Product_Detail' column.")
```

```python
import re

def extract_first_color_from_series(series):
    # Define a regular expression pattern to match colors
    color_pattern = r'[A-Za-z\s]+'
    # Initialize an empty list to store extracted colors
    extracted_colors = []
    # Iterate over each element in the Series
    for line in series:
        # Check if the line is not None
        if line is not None:
            # Check if the line contains any numbers
            if not any(char.isdigit() for char in line):
                # Use regular expression to find the first color in the element
                match = re.search(color_pattern, line)
                if match:
                    # Extract the first color found and remove leading/trailing whitespaces
                    extracted_color = match.group(0).strip()
                    # Append the extracted color to the list
                    extracted_colors.append(extracted_color)
    # Return the list of extracted colors
    return extracted_colors

extracted_colors = extract_first_color_from_series(df['Color_Extracted'])
```

```python
review_list= []

for index, row in df.iterrows():
    asin = row['Asin']
    # Assume 'Reviews' contains string representations of lists of dictionaries
    try:
        reviews_data = ast.literal_eval(row['Reviews'])
    except (ValueError, SyntaxError):
        continue 

    for review in reviews_data:
        # Now 'review' is expected to be a dictionary, so we can use .get()
        stars = review.get('stars', 'Unknown')
        content = review.get('content', 'No review content')
        review_list.append({'Asin': asin, 'Review_Stars': stars, 'Review_Content': content})

# Convert the list of review dictionaries into a DataFrame
reviews_df = pd.DataFrame(review_list)
```

```python
  
#nltk.download('punkt')
#nltk.download('stopwords')
#nltk.download('wordnet')

def clean_text(text):
    # Ensure text is a string; if None, treat it as an empty string
    if text is None:
        return ''
    
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    
    # Remove stop words and non-alphabetic tokens
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    
    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Join the tokens back into a string
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text

# Apply the cleaning function directly to each review in the 'Review_Content' column
reviews_df['Cleaned_Review_Content'] = reviews_df['Review_Content'].apply(clean_text)
```

```python
import os
import requests
import csv

# Function to download images
def download_images(image_urls, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    metadata = []

    for idx, url in enumerate(image_urls, start=1):
        if url is None:
            print(f"Skipping image {idx}: URL is None")
            continue

        try:
            response = requests.get(url)
            if response.status_code == 200:
                # Generate filename
                filename = f"image_{idx}.jpg"
                filepath = os.path.join(output_dir, filename)
                
                # Save image
                with open(filepath, 'wb') as f:
                    f.write(response.content)
                
                metadata.append((idx, filename, url))
                print(f"Downloaded image {idx}")
            else:
                print(f"Failed to download image {idx}: {url}")
        except Exception as e:
            print(f"Error downloading image {idx}: {e}")

    # Write metadata to CSV file
    csv_filepath = os.path.join(output_dir, 'metadata.csv')
    with open(csv_filepath, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Product ID', 'Filename', 'URL'])
        for entry in metadata:
            writer.writerow(entry)
    
    print(f"Metadata saved to {csv_filepath}")

# Download images
download_images(df['Landing_Image'], 'product_images')
```

```python
import re
price_pattern = r'\$[\d.]+'

df['Price_Extracted'] = df['Price_Extracted'].astype(str)
    
# Function to extract last price using regex
def extract_last_price(price_str):
    prices = re.findall(price_pattern, price_str)
    if prices:
        return prices[-1]
    else:
        return None

# Apply function to clean Price_Extracted column
df['Price_Extracted'] = df['Price_Extracted'].apply(extract_last_price)
```

```python
import sqlite3
import pandas as pd

# Create a connection to the database
conn = sqlite3.connect('Amazon_data.db')
cursor = conn.cursor()

cursor.execute('''DROP TABLE IF EXISTS Product''')

# Create the Product table if it does not exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Product (
        Asin TEXT PRIMARY KEY,
        Sub_Category TEXT,
        Sub_Sub_Category TEXT,
        Product_Link TEXT,
        Brand_Name TEXT,
        Product_Title TEXT,
        Cleaned_Product_Detail TEXT,
        Average_Review REAL,
        Color_Extracted TEXT,
        Price_Extracted REAL
    )
''')

def insert_data_from_df_to_db(cursor, df):
    for _, row in df.iterrows():
        # Handle missing values or data inconsistencies
        cleaned_product_detail = row.get('Cleaned_Product_Detail', '')  # Handle missing value for 'Cleaned_Product_Detail'
        product_link = row.get('Product_Link', '')  # Handle missing value for 'Product_Link'
        
        # Insert data into the Product table
        cursor.execute('''
            INSERT INTO Product (
                Asin, Sub_Category, Sub_Sub_Category, Product_Link, 
                Brand_Name, Product_Title, Cleaned_Product_Detail, 
                Average_Review, Color_Extracted, Price_Extracted
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            ''', (
                row['Asin'], row['Sub_Category'], row['Sub_Sub_Category'], 
                product_link, row['Brand_Name'], row['Product_Title'], 
                cleaned_product_detail, row['Average_Review'], 
                row['Color_Extracted'], row['Price_Extracted']
            )
        )
    # Commit the changes after inserting all rows
    conn.commit()

# Call the function to insert data
insert_data_from_df_to_db(cursor, df)

```

```python
def fetch_product_details_with_reviews(cursor, asin):
    cursor.execute('''
    SELECT Product.Asin, Product.Product_Title, Review.Review_Stars, Review.Review_Content 
    FROM Product 
    JOIN Review ON Product.Asin = Review.Asin 
    WHERE Product.Asin = ?
    ''', (asin,))
    rows = cursor.fetchall()
    for row in rows:
        print(f"ASIN: {row[0]}, Title: {row[1]}, Stars: {row[2]}, Review: {row[3]}")

fetch_product_details_with_reviews(cursor, 'B0009XU3H8')
```
